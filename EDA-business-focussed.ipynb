## Importing Libraries 

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Loading business code

```python
biz_large_data = pd.read_csv('biz_large_df.csv')
biz_large_df = pd.DataFrame(biz_large_data)

biz_slim_data = pd.read_csv('biz_slim_df.csv')
biz_slim_df = pd.DataFrame(biz_slim_data)
```

## Business Performance Analysis 

```python
# Analysing Business Performance based on Transaction Volume

# Transaction Volume Analysis
tx_volume_analysis = biz_large_df[['name', 'tx_count']].sort_values(by='tx_count', ascending=False)

# Getting the top 5 and bottom 5 businesses based on transaction count
top_5_tx_volume = tx_volume_analysis.head(5)
bottom_5_tx_volume = tx_volume_analysis.tail(5)


(top_5_tx_volume, bottom_5_tx_volume)
```

```python
# Analysing Business Performance based on Revenue

# Revenue Analysis
revenue_analysis = biz_large_df[['name', 'tx_sum_gbp']].sort_values(by='tx_sum_gbp', ascending=False)

# Getting the top 5 and bottom 5 businesses based on revenue
top_5_revenue = revenue_analysis.head(5)
bottom_5_revenue = revenue_analysis.tail(5)

(top_5_revenue, bottom_5_revenue)
```

```python
# Average Transaction Value Analysis
avg_tx_value_analysis = biz_large_df[['name', 'tx_mean_gbp']].sort_values(by='tx_mean_gbp', ascending=False)

# Getting the top 5 and bottom 5 businesses based on average transaction value
top_5_avg_tx_value = avg_tx_value_analysis.head(5)
bottom_5_avg_tx_value = avg_tx_value_analysis.tail(5)

(top_5_avg_tx_value, bottom_5_avg_tx_value)
```

```python
# Customer Reach Analysis
# Counting the number of unique customers for each business
# Assuming 'customers_sorted_unique' is a string representation of a list, we need to evaluate it
import ast
biz_large_df['unique_customers_count'] = biz_large_df['customers_sorted_unique'].apply(lambda x: len(ast.literal_eval(x)))

customer_reach_analysis = biz_large_df[['name', 'unique_customers_count']].sort_values(by='unique_customers_count', ascending=False)


# Top 5 businesses based on customer reach
top_5_customer_reach = customer_reach_analysis.head(5)
bottom_5_customer_reach = customer_reach_analysis.tail(5)

(top_5_customer_reach, bottom_5_customer_reach)
```

## Data Visualisation

```python
import matplotlib.pyplot as plt

# Setting up the figure and axes for a grid of plots
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Business Performance Analysis', fontsize=16)

# Top 5 businesses by transaction volume
axes[0, 0].barh(top_5_tx_volume['name'], top_5_tx_volume['tx_count'], color='skyblue')
axes[0, 0].set_title('Top 5 Businesses by Transaction Volume')
axes[0, 0].set_xlabel('Transaction Count')
axes[0, 0].set_ylabel('Business Type')

# Top 5 businesses by revenue
axes[0, 1].barh(top_5_revenue['name'], top_5_revenue['tx_sum_gbp'], color='lightgreen')
axes[0, 1].set_title('Top 5 Businesses by Revenue')
axes[0, 1].set_xlabel('Total Revenue (GBP)')
axes[0, 1].set_ylabel('Business Type')

# Top 5 businesses by Mean transaction value
axes[1, 0].barh(top_5_avg_tx_value['name'], top_5_avg_tx_value['tx_mean_gbp'], color='salmon')
axes[1, 0].set_title('Top 5 Businesses by Mean Transaction Value')
axes[1, 0].set_xlabel('Mean Transaction Value (GBP)')
axes[1, 0].set_ylabel('Business Type')

# Top 5 businesses by customer reach
axes[1, 1].barh(top_5_customer_reach['name'], top_5_customer_reach['unique_customers_count'], color='gold')
axes[1, 1].set_title('Top 5 Businesses by Customer Reach')
axes[1, 1].set_xlabel('Unique Customer Count')
axes[1, 1].set_ylabel('Business Type')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# Setting up the figure and axes for a grid of plots
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Business Performance Analysis', fontsize=16)

# Bottom 5 businesses by transaction volume
axes[0, 0].barh(bottom_5_tx_volume['name'], bottom_5_tx_volume['tx_count'], color='skyblue')
axes[0, 0].set_title('Bottom 5 Businesses by Transaction Volume')
axes[0, 0].set_xlabel('Transaction Count')
axes[0, 0].set_ylabel('Business Type')

# Bottom 5 businesses by revenue
axes[0, 1].barh(bottom_5_revenue['name'], bottom_5_revenue['tx_sum_gbp'], color='lightgreen')
axes[0, 1].set_title('Bottom 5 Businesses by Revenue')
axes[0, 1].set_xlabel('Total Revenue (GBP)')
axes[0, 1].set_ylabel('Business Type')

# Bottom 5 businesses by mean transaction value
axes[1, 0].barh(bottom_5_avg_tx_value['name'], bottom_5_avg_tx_value['tx_mean_gbp'], color='salmon')
axes[1, 0].set_title('Bottom 5 Businesses by Mean Transaction Value')
axes[1, 0].set_xlabel('Mean Transaction Value (GBP)')
axes[1, 0].set_ylabel('Business Type')

# Bottom 5 businesses by customer reach
axes[1, 1].barh(bottom_5_customer_reach['name'], bottom_5_customer_reach['unique_customers_count'], color='gold')
axes[1, 1].set_title('Bottom 5 Businesses by Customer Reach')
axes[1, 1].set_xlabel('Unique Customer Count')
axes[1, 1].set_ylabel('Business Type')

plt.tight_layout(rect=[0, 0.03, 1, 0.95]) 
plt.show()
```

```python
# Calculating quartiles for each metric again
tx_count_quartiles = biz_large_df['tx_count'].quantile([0.25, 0.50, 0.75])
revenue_quartiles = biz_large_df['tx_sum_gbp'].quantile([0.25, 0.50, 0.75])
mean_tx_value_quartiles = biz_large_df['tx_mean_gbp'].quantile([0.25, 0.50, 0.75])
median_tx_value_quartiles = biz_large_df['tx_median_gbp'].quantile([0.25, 0.50, 0.75])
customer_reach_quartiles = biz_large_df['unique_customers_count'].quantile([0.25, 0.50, 0.75])

# Function to classify each business based on quartiles
def classify_segment(value, quartiles):
    if value <= quartiles[0.25]:
        return 'Low'
    elif value <= quartiles[0.50]:
        return 'Low-medium'
    elif value <= quartiles[0.75]:
        return 'Upper-medium'
    else:
        return 'High'

# Segmenting businesses
biz_large_df['Transaction Volume Segment'] = biz_large_df['tx_count'].apply(classify_segment, quartiles=tx_count_quartiles)
biz_large_df['Revenue Segment'] = biz_large_df['tx_sum_gbp'].apply(classify_segment, quartiles=revenue_quartiles)
biz_large_df['Mean Transaction Value Segment'] = biz_large_df['tx_mean_gbp'].apply(classify_segment, quartiles=mean_tx_value_quartiles)
biz_large_df['Median Transaction Value Segment'] = biz_large_df['tx_median_gbp'].apply(classify_segment, quartiles=median_tx_value_quartiles)
biz_large_df['Customer Reach Segment'] = biz_large_df['unique_customers_count'].apply(classify_segment, quartiles=customer_reach_quartiles)

# Displaying a sample of the segmented data
segmented_data_sample = biz_large_df[['name', 'Transaction Volume Segment', 'Revenue Segment', 'Mean Transaction Value Segment', 'Median Transaction Value Segment', 'Customer Reach Segment']].head(20)
segmented_data_sample
```
